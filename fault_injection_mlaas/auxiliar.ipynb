## Read dataset

```python
import pandas as pd

tweets = pd.read_csv('Tweets_dataset.csv')
tweets['word_count'] = tweets.text.apply(lambda t: len(t.split()))

tweets.describe()
```

## Plot histogram

```python
import plotly.graph_objects as go

tweets_sentiments_count = tweets.loc[ : , ['airline_sentiment', 'word_count']]

fig = go.Figure()

for sentiment, grouped in tweets_sentiments_count.groupby('airline_sentiment'):
  histogram=go.Histogram(x=grouped.word_count, name=sentiment, bingroup=1, opacity=0.75)
  fig.add_trace(histogram)

fig.add_hline(y=34, line_dash="dash", line_color="green", line_width=1)
fig.add_vrect(x0=3, x1=26, line_width=0, fillcolor="green", opacity=0.2)

fig.update_layout(xaxis = dict(tickmode = 'linear', tick0 = 0, dtick = 1),
                  bargap=0.3,
                  title='Dataset sentence word count distribution per class',
                  xaxis_title="sentence word count", yaxis_title="frequency")

fig.show()
```

Chosed sentence sizes: 10, 18 and 26
