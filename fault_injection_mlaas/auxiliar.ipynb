# Word Count Distribution

```python
%matplotlib widget

import pandas as pd

tweets = pd.read_csv('Tweets_dataset.csv')
tweets['word_count'] = tweets.text.apply(lambda t: len(t.split()))

tweets = tweets.loc[ : , ['airline_sentiment', 'text', 'word_count']]

bins = len(tweets.groupby('word_count'))
tweets.hist(bins=bins)
```
